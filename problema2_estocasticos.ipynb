```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def V(figx=None):
    rho = np.linspace(0, 0.4, 20)

    for l in [4.5, 5.5, 8.5, 15.5]:
        V = (0.25 * l * (rho ** 4)) - (l * rho ** 3) / 3 + (0.5 * rho ** 2)
    
        figx.plot(rho, V, '-', label=(f'$\lambda={l}$'))
        figx.axhline(0, color='k', lw=0.5)
        
        rhopos = 0.5 * (1 + np.sqrt(1 - (4 / l)))
        rhoneg = 0.5 * (1 - np.sqrt(1 - (4 / l)))
        figx.plot(rhoneg, 0, 'ok')
    
fig, fig1 = plt.subplots(1, 1, figsize=(12, 6))

V(figx=fig1)
plt.legend()
```

```python
def rhostar(figx=None):
    l = np.arange(0, 10, 0.5)
    
    rho0 = np.zeros(len(l))
    
    rhopos = 0.5 * (1 + np.sqrt(1 - (4 / l)))
    rhoneg = 0.5 * (1 - np.sqrt(1 - (4 / l)))
    
    rhopos[l==4] = 0.5
    rhoneg[l==4] = 0.5
    
    figx.plot(l[l>=4], rhopos[l>=4], color='k', ls='-', label='Fase activa estable')
    figx.plot(l[l>=4], rhoneg[l>=4], color='k', ls='--', label='Fase activa inestable')
    figx.plot(l, rho0, color='k', ls='-', label='Fase absorbente')
    figx.plot(4, 0.5, 'ok', label='Punto de silla')
    figx.legend()
fig, fig2 = plt.subplots(1, 1, figsize=(12, 6))
rhostar(figx=fig2)

plt.savefig('prob_2_4.PNG')
```
